```python
from sklearn.preprocessing import LabelEncoder

# Daftar kolom yang akan dienkode
categorical_columns = ['Nama','Jenis_Kelamin', 'Kelurahan', 'Kecamatan','Cara_Ukur1','BB_TB1','BB_U1','TB_U1','Pola_Asuh1','Vit_A1','Asi_Eksklusif1',
                       'Cara_Ukur2','BB_TB2','BB_U2','TB_U2','Pola_Asuh2','Vit_A2','Asi_Eksklusif2',
                       'Cara_Ukur3','BB_TB3','BB_U3','TB_U3','Pola_Asuh3','Vit_A3','Asi_Eksklusif3',
                       'Cara_Ukur4','BB_TB4','BB_U4','TB_U4','Pola_Asuh4','Vit_A4','Asi_Eksklusif4',
                       'Cara_Ukur5','BB_TB5','BB_U5','TB_U5','Pola_Asuh5','Vit_A5','Asi_Eksklusif5',
                       'Cara_Ukur6','BB_TB6','BB_U6','TB_U6','Pola_Asuh6','Vit_A6','Asi_Eksklusif6',
                       'Cara_Ukur7','BB_TB7','BB_U7','TB_U7','Pola_Asuh7','Vit_A7','Asi_Eksklusif7']

# Inisialisasi label encoder
label_encoders = {}

# Terapkan label encoding ke setiap kolom kategorikal
for column in categorical_columns:
    le = LabelEncoder()
    df1[column] = le.fit_transform(df1[column])
    label_encoders[column] = le

# Tampilkan beberapa baris pertama dari dataframe yang telah diperbarui
df1.head()
```

```python
# Fungsi untuk membuat directori untuk pengkodean label
def create_label_dict(label_encoder):
    return {index: label for index, label in enumerate(label_encoder.classes_)}

# membuat direktori untuk menyimpan pengkodean label untuk setiap kolom kategorikal
label_encoding_info = {column: create_label_dict(le) for column, le in label_encoders.items()}

label_encoding_info
```
